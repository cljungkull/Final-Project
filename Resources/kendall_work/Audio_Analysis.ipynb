```python
# standard imports
%matplotlib inline
from matplotlib import pyplot as plt
import matplotlib as mpl
import pandas as pd
# found this library when looking at this competiton - it proved worthy! :)

```

```python
! awk 'NR == 1 || NR % 10 == 0' Resources/train_every_Nth.csv > Resources/new_train_data.csv
```

```python
# let's create DataFrame from those lines
audio_df = pd.read_csv("Resources/new_train_data.csv", skiprows=1, names=["sound", "time"])

# let's index every earthquake
quake_indices = audio_df.index[audio_df.time.diff()>0]
quake_indices

# creating figure and clipping dataset
plt.figure(figsize=(10, 5))
data = audio_df.sound.clip(-75,75)

# a simple "data.plot()" would've suffice.. just to make it look a little bit better
for i in range(50):
    data[i::50].plot(alpha=0.05,color="midnightblue")

plt.gca().set_title('Whole dataset - pink dashed lines are earthquakes')

# drawing eartquake lines
for quake in quake_indices:
    plt.axvline(x=quake, linestyle = ":", color="red")
```
